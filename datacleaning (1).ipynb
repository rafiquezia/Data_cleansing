```python
#importing libraries
import pandas as pd
```

```python
#loading the dataset
df = pd.read_excel('D:/DSAI/Data/OnlineRetail.xlsx')
```

```python
#checking the shape of dataset before cleaning
df.shape
data=df.copy()
data1 = data.copy()
data.head()
```

```python
#finding the number of null values in each column
df.isnull().sum()
```

```python
#dropping all the null values in the data
df.dropna(inplace=True)
```

```python
#checking the shape of data after removing the null values
print(df.shape)
#data types of each column
df.dtypes
```

```python
#checking for alphanumeric data in numeric data columns
column_name = 'InvoiceNo'
df[column_name] = df[column_name].astype(str, errors='ignore')
ir = df[column_name].str.contains('[A-Z]', regex=True, na=False)
ir
#we found that there are some alphaumneric values like c581490 
#ir contains all the entries having alphanumeric values
df = df[~ir]
df.head()

```

```python
#now lets look for the irregular or incorrect entries in strings
column = 'StockCode'
df[column] = df[column].astype(str, errors='ignore')

ir = df[column].str.contains('[A-Z]', regex=True, na=False)
#since stockcode contains both numbers and words, removing all the numerical entries
df = df[ir]
df.head()
```

```python
#stockcode has a unique format, removing all the entries not foolowing the format
cn = 'StockCode'
valid = df[cn].str.match(r'\d{5}[A-Z]')
df=df[valid]
df.shape
```

```python
#checking for irregular numerics in numerical data based on the length of entry
numeric_column = 'Country'
df[numeric_column] = df[numeric_column].astype(str)
df['length'] = df[numeric_column].str.len()

# Filter rows based on the length condition
filtered_df = df[df['length'] < 5]
filtered_df

# Drop the length column if it's no longer needed
df.drop('length', axis=1, inplace=True)
```

```python
#changing the data types before saving the data
df['InvoiceNo'] = df['InvoiceNo'].astype(int)
df['Quantity'] = df['Quantity'].astype(int)
df['UnitPrice'] = df['UnitPrice'].astype(float)
df['CustomerID'] = df['CustomerID'].astype(float)
df['Country'] = df['Country'].astype(str)
df['CustomerID'] = df['CustomerID'].astype(int)
df.drop('Stockcode', axis=1, inplace=True)
df.dtypes
df.head()
```
